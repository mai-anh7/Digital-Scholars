4. Data Science

Since video category isn't a numerical value, one question that I have about my dataset is if the channel type rank is correlated to the amount of video views. To answer this question, I plan on using python to the correlation coefficient between these two varibles. Specifically, I'm going to use the .corr function to do so.

```python
category_mean.corr(numeric_only=True)['video views'].sort_values(ascending=False)
category_mean.plot.scatter(x = "channel_type_rank", y = "video views", title = "Average Video Views among Youtube Videos of Different Channel Type Rankings")
#correlation coefficient between video views & channel type rank = -0.083612
#negative correlation
#not perfect correlation since the correlation coefficient isn't exactly 1 or -1 so we cannot perfectly predict the amount of video views by knowing the channel ranking
#Since this correlation coefficient is closer to 0, there may not be any correlation/association between the two variables
#not casuation
#however, since the amount of video views the shows video category has is demonstrated as an outlier (as seen in the graph, I decided to remove it from the graph and see if it had an impact on the correlation efficient between channel type rank and video views)

mean_noShows = category_mean[category_mean["category"] != "Shows"]
mean_noShows.corr(numeric_only=True)['video views'].sort_values(ascending=False)
mean_noShows.plot.scatter(x = "channel_type_rank", y = "video views", title = "Average Video Views among Youtube Videos of Different Channel Type Rankings")
#correlation coefficient between channel type rank & video views  w/o Shows = 0.2
```

5. Overall summary

For your information, my dataset is one that contains data on some of the top global Youtube statistics of 2023, including each video's category, channel rank, title, subscribers, etc. To begin my project, I grouped each video category together in order to find the average video views for all the videos apart of each category. Once that was completed, I plotted my data on a scatter plot graph to better visualize my data. After doing so and reviewing my data table, it became apparent that the "shows" category was the video category with the most video views by far. To emphasize this, the "shows" video category had 34 billion views, with the next closest video category at 17 billion. I did the same process to find the median video views and found that the "trailers" category had the greatest median at 17 billion while the "shows" category trailed in close second at 14 billion.


After analyzing all of this data, especially the mean scatter plot graph, a question came to me: Is channel type ranking correlated with the number of video views? I decided to find the correlation coefficient since it measures the strength of the linear association between the x and y variables. After finding the correlation coefficient between these two variables, I found that there was negative correlation as the correlation coefficient was -0.08. This meant that that there was little correlation since this number is closer to 0 than 1 or -1. However, with the knowledge that the "shows" catgeory was at outlier with its amount of video views being 17 billion more than the category second to it, I decided that it was having too great of an effect on the correlation coefficient and that the reason for this may have been false data. After finding the correlation coefficient again, this time without the video views for the "shows" category, the correlation coefficient went up to 0.2. This number most likely signifies that there is little correlation between channel type ranking and the number of video views because again, 0.2 is close to 0 and 0 means there is no correlation.
